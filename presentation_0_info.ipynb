# Simulador MTLK_TR1



## Introduo

Este trabalho foca na simulao do funcionamento das camadas de enlace e fsica, atravs da implementao de protocolos de enquadramento, modulao banda-base e modulao por portadora. Com o propsito de aprofundar o entendimento dos protocolos das camadas de enlace de dados e fsica, o estudo  dividido em etapas, incluindo simulaes de modulaes digitais como NRZ-Polar, Manchester, Bipolar, ASK, FSK e 8-QAM, alm da implementao de protocolos de enquadramento de dados e deteco e correo de erros.

Uma explicao detalhada do funcionamento do simulador pode ser encontrada nos arquivos `presentation_n.pdf`, que descrevem cada camada implementada

## Execuo

Para executar o simulador, basta executar o arquivo `app.py`.

## Dependncias

Para instalar as dependncias, basta executar o comando `pip install -r requirements.txt` na raiz do projeto.

## Concluso

O desenvolvimento deste trabalho proporcionou um entendimento prtico e detalhado dos protocolos das camadas de enlace e fsica em redes de computadores. Atravs da implementao e simulao dos diversos protocolos, foram evidenciadas as complexidades e desafios na transmisso de dados. As dificuldades encontradas ao longo do trabalho reforam a importncia de um entendimento aprofundado dessas camadas para o desenvolvimento e manuteno de sistemas de comunicao eficientes e confiveis.
